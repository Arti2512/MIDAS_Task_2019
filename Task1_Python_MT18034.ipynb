```python
import tweepy as tp
import pandas as pd
import jsonlines
```

# Init Twitter access credentials

```python
def init_access():
    t_id=''
    t_secret=''
    a_id=''
    a_secret=''
    return t_id,t_secret,a_id,a_secret
```

# Authorization

```python
#Reference: https://www.geeksforgeeks.org/extraction-of-tweets-using-tweepy/

def twitter_user_authorize(u_id,u_key,access_id,access_key):
    auth = tp.OAuthHandler(u_id,u_key) 
    auth.set_access_token(access_id, access_key) 
    api = tp.API(auth) 
    return api
```

```python
#Refernce:http://docs.tweepy.org/en/v3.4.0/cursor_tutorial.html
import datetime
def get_tweets(username):
    tmp={}
 
    for status in tp.Cursor(api.user_timeline, screen_name=username, tweet_mode="extended",include_entities=True).items():
        #id = api.get_status(tweet_id)
        dt=status.created_at
        datetimestr = dt.strftime("%H:%M:%S.%f - %b %d %Y") 
        co=0
        c=status.id
        if 'media' in status.entities:
            for image in status.entities['media']:
                if image['type']=='photo':
                    co+=1
            co=str(co)
        if co=='0':
            co='None'
        tmp[c]={'date_and_time':datetimestr,'retweet':status.retweet_count,'favourite_count':status.favorite_count,'text':status.full_text,'No_of_images':co}
   
    print(len(tmp))
    return tmp
```

```python
user_id,user_key,access_id,access_key=init_access()
api=twitter_user_authorize(user_id,user_key,access_id,access_key)
temp=get_tweets('@midasIIITD')
tt=pd.DataFrame(temp)
out_dict=tt.T
```
